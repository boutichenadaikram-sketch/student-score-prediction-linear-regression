```python
#as we said before pandas reads data 
data = pd.read_csv("C:\\Users\\ideap\\Desktop\\student-score-prediction-linear-regression\\data\\student_scores.csv")

```

```python
#name the feature x :all values x : array x 
x = data['Hours'].values
#name the label y : actuale y  : array y 
y = data['Scores'].values
print(x)
print(y)
print(x.shape) # (m*n) (rows*columns) (العمود الصف )
print(y.shape)
```

```python
#FIRST STEP LINEAR REGRESSION : VESUALISE THE DATA 
#we use matplotlib library , contains plt function : 
plt.figure(figsize=(8,6)) #size of picture (graph /figure)
plt.title('Data distribution') #title of figure
plt.scatter(x, y)              #relation between x&y (graph x,y) type of fegure : scatter
plt.xlabel('hours')            #name of mi7war x 
plt.ylabel('score')   #name of mi7war y 
plt.savefig("results/plot1_DataDistribution.png")   #to save the plot in result folder 
plt.show()                     #
```

```python
#SECOND : CHOOSE THE LEARNING MODAL 
#from the figure we see that the relationship between feature x and y is linear 
#so the modal that we are going to choose is linear regression modal 
#RECALL:MATH
```

```python
#THIRD : THE MODAL 
#THERD_1 : Before the learning 
# Hyper-parameteres
L_rate = 0.001
iterations = 100

# Initialization
theta_1= 0
theta_0= 0

# The number of samples in the dataset
n = x.shape[0]
# An empty list to store the error in each iteration(values of MSE)
losses = []
```

```python
#THIRD_2 : START THE LEARNING 
for i in range(iterations):
    h_x = theta_0 + theta_1*x
    
    # Keeping track of the error decrease
    mse = (1/n) * np.sum((h_x - y)**2)
    losses.append(mse)
    
    # Derivatives
    d_theta0 = (2/n) * np.sum(h_x-y)
    d_theta1 = (2/n) * np.sum(x * (h_x-y))
    
#     # Values update (Gradient descent)
    theta_1 = theta_1 - L_rate * d_theta1
    theta_0 = theta_0 - L_rate* d_theta0
```

```python
print("theta_0= ", theta_0)
print("theta_1= ", theta_1)
print("MSE= ", mse)
```

```python
#result : prediction modal (we are going to use it to predict y of new x ) 
# H(x)=1.63 + 9.89 x 
```

```python
#lets predict score for new x (new hours) : 
new_x= 9

Prediction_Model = theta_0 + theta_1*new_x
print ('Score:', Prediction_Model)
```

```python
plt.title('Loss values vs iterations')
plt.plot(losses)
plt.ylabel('loss')
plt.xlabel('iteration')
print('Initial loss\t:', losses[0])
print('Final loss\t:', losses[-1])
plt.savefig("results/plot2_LossvaluesVsIterations.png")
```
